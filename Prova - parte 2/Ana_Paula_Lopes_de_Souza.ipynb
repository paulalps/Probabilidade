## 6) Existe correlação entre a nota do enem e o status do aluno?

```python
enem_ativo = dados.loc[dados['status'].isin(['ATIVO'])]
enem_cancelado = dados.loc[dados['status'].isin(['CANCELADO'])]

media_ativo = enem_ativo['enen-nota'].mean()
media_cancelado = enem_cancelado['enen-nota'].mean()

print('A média das notas do ENEM dos alunos com vínculo ativo é de ', media_ativo)
print('A média das notas do ENEM dos alunos com vínculo cancelado é de ', media_cancelado)
```

## Não há correlação entre as notas do enem e o status do aluno.

## 7)Existe algum componente curricular na base de dados que possui uma alta correlação com os dados do ENEN?
