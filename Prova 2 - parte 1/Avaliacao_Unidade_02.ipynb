````python
import pandas as pd

turmas = pd.read_csv("turmas-2017.1.csv", sep = ";")
turmas_imd = turmas[turmas['codigo'].str.contains("IMD")]
turmas_imd_cons = turmas_imd[turmas_imd['situacao_turma'].str.contains("CONSOLIDADA")]
turmas_imd_cons = turmas_imd_cons[~turmas_imd_cons.duplicated('id_turma')]
print('1. Foram consolidadas ' + str(len(turmas_imd_cons)) + ' turmas:')
print(turmas_imd_cons['codigo'].unique())
````

## (3 pontos) Realize um estudo envolvendo as propriedades de assimétria e curtose para as turmas IMD0033 (probabilidade) consolidadas em 2017.1. Utilize gráficos e explique os resultados. Reforce também os horários das disciplinas nos resultados. 

````python
import matplotlib.pyplot as plt
import seaborn as sns
#Assimetria
from scipy.stats import skew
#Curtose
from scipy.stats import kurtosis

fig, (ax1, ax2) = plt.subplots(figsize=(10,5), ncols=2, nrows=1)

probabilidade = turmas_imd_cons[turmas_imd_cons['codigo'].str.contains("IMD0033")]
id_probabilidade = list(probabilidade['id_turma'])

matricula_comp_id_turma1 = matricula_comp[matricula_comp['id_turma'] == id_probabilidade[0]]
matricula_comp_id_turma2 = matricula_comp[matricula_comp['id_turma'] == id_probabilidade[1]]

turma1 = matricula_comp_id_turma1.dropna()
turma2 = matricula_comp_id_turma2.dropna()

assimetria_turma1 = skew(turma1['nota'])
curtose_turma1 = kurtosis(turma1['nota'])

assimetria_turma2 = skew(turma2['nota'])
curtose_turma2 = kurtosis(turma2['nota'])

# main title
plt.suptitle("Curtose", fontsize=20, y = 1.09)

### Titles of subplots
ax1.set_title("Turma 1(kurtosis = " + str(curtose_turma1) + ")", fontsize=12)
ax2.set_title("Turma 2(kurtosis = " + str(curtose_turma2) + ")", fontsize=12)

sns.distplot(turma1['nota'], kde = False, ax=ax1, bins=10)
sns.distplot(turma2['nota'], kde = False, ax=ax2, bins=10)

plt.show()
````

````python
fig1, (ax3, ax4) = plt.subplots(figsize=(10,5), ncols=2, nrows=1)

# organize space among figures
fig1.tight_layout()

# main title
plt.suptitle("Assimetria", fontsize=20, y = 1.09)

### Titles of subplots
ax3.set_title("Turma 1(skew = " + str(assimetria_turma1) + ")", fontsize=12)
ax4.set_title("Turma 2(skew = " + str(assimetria_turma2) + ")", fontsize=12)

sns.distplot(turma1['nota'], kde = False, ax=ax3, bins=10)
sns.distplot(turma2['nota'], kde = False, ax=ax4, bins=10)

print('Valores - assimetria',
      '\nTurma 1:', skew(turma1['nota']),
      '\nTurma 2: ', skew(turma2['nota']),)

plt.show()
````

````python
fig4, (ax_edb1, ax_edb2, ax_edb3, ax_edb4) = plt.subplots(figsize=(10,5), ncols=4, nrows=1)

fig4.tight_layout()
# main title
plt.suptitle("Media (vermelho) vs Mediana (verde)", fontsize=20, y = 1.09)

y_title_margin = 1

### Titles of subplots
ax_edb1.set_title("Turma 1", y = y_title_margin, fontsize=12)
ax_edb2.set_title("Turma 2", y = y_title_margin, fontsize=12)
ax_edb3.set_title("Turma 3", y = y_title_margin, fontsize=12)
ax_edb4.set_title("Turma 4", y = y_title_margin, fontsize=12)

# Plot the mean in red
ax_edb1.axvline(turma1_edb_1['media_final'].mean(), color="r")
ax_edb2.axvline(turma2_edb_1['media_final'].mean(), color="r")
ax_edb3.axvline(turma3_edb_1['media_final'].mean(), color="r")
ax_edb4.axvline(turma4_edb_1['media_final'].mean(), color="r")

# Plot the median in green
ax_edb1.axvline(turma1_edb_1['media_final'].median(), color="g")
ax_edb2.axvline(turma2_edb_1['media_final'].median(), color="g")
ax_edb3.axvline(turma3_edb_1['media_final'].median(), color="g")
ax_edb4.axvline(turma3_edb_1['media_final'].median(), color="g")

sns.distplot(turma1_edb_1['media_final'], kde = False, ax=ax_edb1,bins=20)
sns.distplot(turma2_edb_1['media_final'], kde = False, ax=ax_edb2,bins=20)
sns.distplot(turma3_edb_1['media_final'], kde = False, ax=ax_edb3,bins=20)
sns.distplot(turma4_edb_1['media_final'], kde = False, ax=ax_edb4,bins=20)


print('Valores',
      '\nTurma 1: ', skew(turma1_edb_1['media_final']),
      '\nTurma 2: ', skew(turma2_edb_1['media_final']),
      '\nTurma 3: ', skew(turma3_edb_1['media_final']),
      '\nTurma 4: ', skew(turma4_edb_1['media_final']))
      

    
    
plt.show()
````
